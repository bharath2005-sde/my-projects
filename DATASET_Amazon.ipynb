```python
import zipfile
import pandas as pd

# Unzip the file
with zipfile.ZipFile('/content/amazon.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/')

# Check extracted files
import os
print(os.listdir('/content'))

# Load CSV (change name if needed)
df = pd.read_csv('/content/amazon.csv')

# Show first 5 rows
df.head()
```

```python
df.head()
df.describe()
df.isnull().sum()
```

```python
# Drop duplicates
df = df.drop_duplicates()

# Fill missing values (example)
df = df.fillna(method='ffill')

# Check again
df.isnull().sum()
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip().str.lower()
print(df.columns)
```
